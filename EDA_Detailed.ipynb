# Removal of Outliers

```python
data = pd.get_dummies(data)
dummy_data = data.copy()
data.boxplot( rot = 90, figsize=(20,5))
```

```python
var = 'viscera_wt'
plt.scatter(x = data[var], y = data['age'],)
plt.grid(True)
```

```python
data.drop(data[(data['viscera_wt']> 0.5) & (data['age'] < 20)].index, inplace=True)
data.drop(data[(data['viscera_wt']<0.5) & (data['age'] > 25)].index, inplace=True)
```

```python
var = 'shell_wt'
plt.scatter(x = data[var], y = data['age'],)
plt.grid(True)
```

```python
data.drop(data[(data['shell_wt']>=0.8) & (data['age'] < 20)].index, inplace=True)
data.drop(data[(data['shell_wt']<0.6) & (data['age'] > 20)].index, inplace=True)
```

```python
var = 'shucked_wt'
plt.scatter(x = data[var], y = data['age'],)
plt.grid(True)
```

```python
data.drop(data[(data['shucked_wt']>= 1.2) & (data['age'] < 20)].index, inplace=True)
data.drop(data[(data['shucked_wt']<1) & (data['age'] > 20)].index, inplace=True)
```

```python
var = 'whole_wt'
plt.scatter(x = data[var], y = data['age'],)
plt.grid(True)
```

```python
data.drop(data[(data['whole_wt']>= 2.5) & (data['age'] < 25)].index, inplace=True)
data.drop(data[(data['whole_wt']<2.5) & (data['age'] > 25)].index, inplace=True)
```

```python
var = 'height'
plt.scatter(x = data[var], y = data['age'],)
plt.grid(True)
```

```python
data.drop(data[(data['height']>0.4) & (data['age'] < 15)].index, inplace=True)
data.drop(data[(data['height']<0.4) & (data['age'] > 25)].index, inplace=True)
```

```python
var = 'length'
plt.scatter(x = data[var], y = data['age'],)
plt.grid(True)
```

```python
data.drop(data[(data['length']<0.1) & (data['age'] < 5)].index, inplace=True)
data.drop(data[(data['length']<0.8) & (data['age'] > 25)].index, inplace=True)
data.drop(data[(data['length']>=0.8) & (data['age']< 25)].index, inplace=True)
```

```python
var = 'diameter'
plt.scatter(x = data[var], y = data['age'],)
plt.grid(True)
```

```python
data.drop(data[(data['diameter']<0.1) & (data['age'] < 5)].index, inplace=True)
data.drop(data[(data['diameter']<0.6) & (data['age'] > 25)].index, inplace=True)
data.drop(data[(data['diameter']>=0.6) & (data['age']< 25)].index, inplace=True)
```
